```python
def plotLossCdlc(fname,head,size,step,show):
    lossVal = mres.LossValues()
    print(fname)
    if os.path.isfile(fname) == False:
        print("file not found - %s " % fname)
        return
    lossVal = pickle.load( open( fname, "rb" ) )
    lenVal = len(lossVal.x)
    lnt=lenVal
    if size != 0:
        lnt = size
    mn = min(np.array(lossVal.y[0:lnt-1]))
    mx = max(np.array(lossVal.y[0:lnt-1]))
    rr = list(range(0,lnt,step))
    x = [ list(lossVal.x)[ix] for ix in rr ]
    y = [ lossVal.y[ix] for ix in rr ]
    mean = [ lossVal.mean[ix] for ix in rr ]
    if show == True:
        plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
        plt.plot(np.array(x), np.array(y))
        plt.plot(np.array(x), np.array(mean))
        plt.legend(['err','mean'], loc='upper right')
        plt.title(head)
        plt.xlabel('epoch')
        plt.ylabel('loss')
        plt.show()
    else:
        print("%d %d %d" % (len(lossVal.x),len(lossVal.y),len(lossVal.mean)))
```

```python
def plotStatsCdlc(fname,head,size,show):
    statsVal = {}
    statsValTrn = mres.ModelStats()
    statsValTst = mres.ModelStats()
    print(fname)
    if os.path.isfile(fname) == False:
        print("file not found - %s " % fname)
        return
    statsVal = pickle.load( open( fname, "rb" ) )
    statsValTrn = statsVal["train"]
    statsValTst = statsVal["test"]
    lnt = len(statsValTrn.epoch)
    if size != 0:
        lnt = size
    if show == True:
        plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
        plt.plot(np.array(statsValTrn.epoch[0:lnt-1]), np.array(statsValTrn.accuracy[0:lnt-1]))
        plt.plot(np.array(statsValTrn.epoch[0:lnt-1]), np.array(statsValTrn.errorRate[0:lnt-1]))
        plt.plot(np.array(statsValTrn.epoch[0:lnt-1]), np.array(statsValTrn.f1Score[0:lnt-1]))
        plt.legend(['acc', 'err', 'f1'], loc='upper left')
        plt.title("train " + head)
        plt.xlabel('epoch')
        plt.ylabel('value')
        plt.show()
        plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
        plt.plot(np.array(statsValTst.epoch[0:lnt-1]), np.array(statsValTst.accuracy[0:lnt-1]))
        plt.plot(np.array(statsValTst.epoch[0:lnt-1]), np.array(statsValTst.errorRate[0:lnt-1]))
        plt.plot(np.array(statsValTst.epoch[0:lnt-1]), np.array(statsValTst.f1Score[0:lnt-1]))
        plt.legend(['acc', 'err', 'f1'], loc='upper left')
        plt.title("test " + head)
        plt.xlabel('epoch')
        plt.ylabel('value')
        plt.show()
```

```python
def statsCdlc(fname):
    def posList(lst,val):
        return [i for i in range(len(lst)) if lst[i] == val]
    def epoch(inpList1,inpList2):
        maxVal = max(inpList2)
        return [maxVal,inpList1.epoch[inpList2.index(maxVal)]]
    statsVal = pickle.load( open( fname, "rb" ) )
    statsValTrn = statsVal["train"]
    statsValTst = statsVal["test"]
    
    maxTrn = epoch(statsValTrn,statsValTrn.accuracy)
    maxTst = epoch(statsValTst,statsValTst.accuracy)
    accList = {"train max" : maxTrn[0],"train mean" : np.mean(statsValTrn.accuracy),"train epoch" : maxTrn[1],"posList":posList(statsValTrn.accuracy,maxTrn[0]),
               "test max" : maxTst[0],"test mean" : np.mean(statsValTst.accuracy),"test epoch" : maxTst[1],"posList" : posList(statsValTst.accuracy,maxTst[0])}

    maxTrn = epoch(statsValTrn,statsValTrn.errorRate)
    maxTst = epoch(statsValTst,statsValTst.errorRate)
    errList = {"train max" : maxTrn[0],"train mean" : np.mean(statsValTrn.errorRate),"train epoch" : maxTrn[1],"posList":posList(statsValTrn.accuracy,maxTrn[0]),
               "test max" : maxTst[0],"test mean" : np.mean(statsValTrn.errorRate),"test epoch" : maxTst[1],"posList" : posList(statsValTst.accuracy,maxTst[0])}

    maxTrn = epoch(statsValTrn,statsValTrn.f1Score)
    maxTst = epoch(statsValTst,statsValTst.f1Score)
    f1List = {"train max" : maxTrn[0],"train mean" : np.mean(statsValTrn.f1Score),"train epoch" : maxTrn[1],"posList":posList(statsValTrn.accuracy,maxTrn[0]),
               "test max" : maxTst[0],"test mean" : np.mean(statsValTrn.f1Score),"test epoch" : maxTst[1],"posList" : posList(statsValTst.accuracy,maxTst[0])}

    return {"Accuracy":accList,"Error Rate":errList,"F1 Score": f1List}
```

```python
def strFoldStatsCdlc(fname):
    statsVal = pickle.load( open(fname, "rb" ) )
    fold = len(statsVal['train'].fold)
    vals = statsVal['train']
    result = {}
    strTrain = "precision:%.3f recall:%.3f f1-score:%.3f  score:%.3f accuracy:%.3f error-rate:%.3f " % \
    (sum(vals.precision)/fold,sum(vals.recall)/fold,sum(vals.f1Score)/fold,sum(vals.score)/fold,\
     sum(vals.accuracy)/fold,sum(vals.errorRate)/fold)
    result["train"] = strTrain
    fold = len(statsVal['test'].fold)
    vals = statsVal['test']
    strTest = ("precision:%.3f recall:%.3f f1-score:%.3f  score:%.3f accuracy:%.3f error-rate:%.3f " % \
      (sum(vals.precision)/fold,sum(vals.recall)/fold,sum(vals.f1Score)/fold, \
      sum(vals.score)/fold,sum(vals.accuracy)/fold,sum(vals.errorRate)/fold))
    result["test"] = strTest
    return result
```

```python
head = ''
size=0
step=1000
classes = ['art','arts','biology','business','creativity','culture','design','economics','education',
           'entertainment','health','politics','science','technology']

for classname in classes:
    fname = 'cdlc-'+ classname +'-additive-minmax-mlp-simple-loss-values-RMSprop-0.001-0.5-300-10-2.bin'
    head = fname[0:fname.find("mlp-simple")-1]
    plotLossCdlc(fname,head,size,step,True)

    fname = 'cdlc-'+ classname +'-tanh-minmax-mlp-simple-loss-values-RMSprop-0.001-0.5-300-10-2.bin'
    head = fname[0:fname.find("mlp-simple")-1]
    plotLossCdlc(fname,head,size,step,True)
```

```python
classes = ['art','arts','biology','business','creativity','culture','design','economics','education',
           'entertainment','health','politics','science','technology']
additive_minmax = {}
tanh_minmax= {}
foldsStatAdd= {}
foldsStatTanh= {}
for classname in classes:
    fname ="cdlc-"+ classname +"-additive-minmax-mlp-simple-stat-values-RMSprop-0.001-0.5-300-10-2.bin"
    additive_minmax[classname] = statsCdlc(fname)
    foldsStatAdd[classname]= strFoldStatsCdlc(fname)
    fname ="cdlc-"+ classname +"-tanh-minmax-mlp-simple-stat-values-RMSprop-0.001-0.5-300-10-2.bin"
    tanh_minmax[classname] = statsCdlc(fname)
    foldsStatTanh[classname]= strFoldStatsCdlc(fname)
```

```python
for classname in classes:
    addtv = additive_minmax[classname]
    tanh = tanh_minmax[classname]
    print("classname : "+classname+" - Additive-Minmax (2 Hidden Layers)")
    print("            Train            Test")
    print("            mean  - max      mean -  max")
    print("accuracy    %.3f - %.3f     %.3f - %.3f" % (addtv["Accuracy"]["train mean"],addtv["Accuracy"]["train max"],addtv["Accuracy"]["test mean"],addtv["Accuracy"]["test max"]))
    print("error rate  %.3f - %.3f     %.3f - %.3f" % (addtv["Error Rate"]["train mean"],addtv["Error Rate"]["train max"],addtv["Error Rate"]["test mean"],addtv["Error Rate"]["test max"]))
    print("f1 score    %.3f - %.3f     %.3f - %.3f" % (addtv["F1 Score"]["train mean"],addtv["F1 Score"]["train max"],addtv["F1 Score"]["test mean"],addtv["F1 Score"]["test max"]))
    print("\n")
    print("10 folds avgs (train) : %s" % foldsStatAdd[classname]["train"])
    print("10 folds avgs (test)  : %s" % foldsStatAdd[classname]["test"])
    print("\n\n")
    print("classname : "+classname+" - Tanh-Minmax (2 Hidden Layers)")
    print("            Train            Test")
    print("            mean  - max      mean -  max")
    print("accuracy    %.3f - %.3f     %.3f - %.3f" % (tanh["Accuracy"]["train mean"],tanh["Accuracy"]["train max"],tanh["Accuracy"]["test mean"],tanh["Accuracy"]["test max"]))
    print("error rate  %.3f - %.3f     %.3f - %.3f" % (tanh["Error Rate"]["train mean"],tanh["Error Rate"]["train max"],tanh["Error Rate"]["test mean"],tanh["Error Rate"]["test max"]))
    print("f1 score    %.3f - %.3f     %.3f - %.3f" % (tanh["F1 Score"]["train mean"],tanh["F1 Score"]["train max"],tanh["F1 Score"]["test mean"],tanh["F1 Score"]["test max"]))
    print("\n")
    print("10 folds avgs (train) : %s" % foldsStatTanh[classname]["train"])
    print("10 folds avgs (test)  : %s" % foldsStatTanh[classname]["test"])
    print("\n")
    print("--------------------------------------------------------------------------------------------")
    print("\n")
```
